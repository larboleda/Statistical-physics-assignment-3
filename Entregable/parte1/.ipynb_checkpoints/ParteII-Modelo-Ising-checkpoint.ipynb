```python
import numpy as np
import matplotlib.pyplot as plt
import random 
#from numpy.random import rand


%matplotlib inline
```

```python
#--------------------------------------
# Generación de todos los microestados 
# Posibles para un sistema de L átomos
# L = N x N
#--------------------------------------

L = 4*4

n_micro = 2**L

all_microstates = [n_micro]

m = 2*np.random.randint(2,size=(1,L))-1

all_microstates[0] = m

for i in range(1,n_micro):
    
    flag = 0
    
    while (flag == 0):   
        
        m = 2*np.random.randint(2,size=(1,L))-1
        
        for j in range(0,i):
            
            if (m == all_microstates[j]).all(): 
                flag = 0
                break 
        
            else: flag = 1 
                
    all_microstates.append(m)
    
    #print(all_microstates[i])        
```

```python
#--------------------------------------
# Estado inicial y su energía asociada 
#--------------------------------------

def all_microstates_energy(all_microstates):
    Etot = E = 0.0
    Ei = Eij = 0.0
    Ei = []
    for i in range(n_micro):
        for j in range(L):
            
            #valor de los spin vecinos de cada átomo en el microestado [i]
            #con condiciones de forntera periodicas
            ngb1 = all_microstates[i][0][(j+1)%L]
            ngb2 = all_microstates[(i-1)%n_micro][0][j]
            ngb3 = all_microstates[i][0][(j-1)%L]
            ngb4 = all_microstates[(i+1)%n_micro][0][j]
            
            #print("primer vecino tiene espín %d de átomo %d"%(ngb1,j))
            Eij =  all_microstates[i][0][j]*(ngb1 + ngb2 + ngb3 + ngb4)
            
            E +=  Eij
            
        Ei.append(E)
           
        Etot += E    
    
    return Ei, Etot


Ei, Etot = all_microstates_energy(all_microstates)

#--------------------------------------------
# Histograma de energias de los microestados
#--------------------------------------------

n_bins = 100

dim = np.sqrt(L)

OmegaE, extrems = np.histogram(Ei, bins=n_bins, normed=False)

plt.figure(figsize=(10,8))
plt.hist(Ei, bins = n_bins, histtype='bar', color ="k", alpha=0.4, lw = 1.8,label =r"Histograma %d$\times$%d"%(dim,dim))
plt.xlabel(r"$E$")
plt.ylabel(r"$\Omega(E)$")
plt.legend()
plt.grid()
plt.savefig("images/directo/histograma_%dx%d.png"%(dim,dim))
plt.show()

```

```python
#---------------------------------------------------------
# Función para generar un estado de la malla arealtorio:
# cada átomo en un estado de espin aleatorio
#---------------------------------------------------------

def initial_microstate(N):   
    #microestado aleatorio de L átomos. L = N x N
    state = 2*np.random.randint(2, size=(N,N))-1
    return state

#---------------------------------------------------------
# Movidas de monte carlos usando el algoritmo Metropolis 
# para generar o no un microestado nuevo
#---------------------------------------------------------

def monte_carlo_realization(microstate,beta):
   
    for i in range(N):
        for j in range(N):
            
                #escogemos aleatoriamente la posición del 
                #átomo (s) en la malla: (fila(a),columna(b))
                
                a = np.random.randint(0, N) #0 y N-1
                b = np.random.randint(0, N)
               
                s =  microstate[a, b]
                
                #vecinos del átomo s:
                ngb1 = microstate[a,(b+1)%N]
                ngb2 = microstate[(a-1)%(N),b]
                ngb3 = microstate[a,(b-1)%N]
                ngb4 = microstate[(a+1)%(N),b]
            
                deltaE = 2*s*(ngb1 + ngb2 + ngb3 + ngb4)
                
                if deltaE < 0: #acepte con probabilidad igual 1 el cambio de espín
                    s *= -1
                elif random.uniform(0.0,1.0) < np.exp(-deltaE*beta):
                    s *= -1
                    
                #cambie el espín del átomo y guardelo en su posición, 
                #esto genera un nuevo microestado
                microstate[a, b] = s
                
    return microstate

#----------------------------
# Energía de un microestado
#----------------------------

def microstate_energy(microstate):
    energy = 0
    for i in range(len(microstate)):
        for j in range(len(microstate)):
            #S: espín del átomo en la fila i columna j
            S = microstate[i,j]
            ngb1 = microstate[i,(j+1)%N]
            ngb2 = microstate[(i-1)%N, j]
            ngb3 = microstate[i,(j-1)%N]
            ngb4 = microstate[(i+1)%N, j]
            energy += -S*(ngb1 + ngb2 + ngb3 + ngb4)     
    return energy

#---------------------------------
# Magnetización de un microestado
#---------------------------------

def magnetization(microstate):
    mag = np.sum(microstate)
    return mag

#---------------------------------------------
#conteo de microestados con una cierta energía
#---------------------------------------------

def omegaE(energies)
```
